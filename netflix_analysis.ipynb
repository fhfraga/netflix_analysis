A [Netflix](https://www.netflix.com/) é uma empresa de produção e serviço de streaming por assinatura americana. Lançado em 29 de agosto de 1997, oferece uma biblioteca de filmes e séries de televisão por meio de acordos de distribuição, bem como produções próprias, conhecidas como Netflix Originals.

Neste projeto iremos analisar os principais dados financeiros fornecidos trimestralmente pela Netflix entre 2011 e 2021. Os dados foram extraídos da página [Netflix Investor](https://ir.netflix.net/ir-overview/profile/default.aspx)

## DICIONÁRIO DE DADOS

| Variável| Descrição|
|:--------|:---------|
| **Data**| É a data de emissão do resultado trimestral|
| **total_de_assinatura_no_final_do_periodo** | É o número total de assinaturas registradas no final do trimestre (*em milhões de assinantes*)|
| **assinaturas_pagas_no_final_do_periodo**| É o número de assinaturas pagas registradas no final do trimestre (*em milhões de assinantes*)|
| **assinaturas_gratuitas**| É o número de assinaturas em período de utilização gratuita registradas no final do trimestre (*em milhões de assinantes*)|
| **receita**| É a receita trimestral gerada para o *Streaming* mundial (*em milhares de dólares*)|
| **custos_da_receita**| É o custo trimestral da receita gerada para o *Streaming* mundial (*em milhares de dólares*)|
| **custo_de_marketing**| É o custo trimestral de Marketing gerado para o *Streaming* mundial (*em milhares de dólares*)|
| **lucro** | É o lucro da contribuição trimestral mundial (*em milhares de dólares*)|
| **margem_de_lucro** | É a margem de lucro trimestral mundial (*em milhares de dólares*)|
| **custo_por_cliente_(excluindo_marketing)**| É a taxa de custo por cliente desconsiderando os gastos com Marketing (*em milhares de dólares*)|
| **receita_por_cliente**  | É a taxa de receita por cliente (*em milhares de dólares*)|
| **lucro_por_cliente** | É a taxa de lucro por cliente (*em milhares de dólares*)|
| **segmento**| É o segmento de operação da companhia ao qual os dados se referem|

## ESTATÍSTICA DESCRITIVA

```python
num_attributes = dados.select_dtypes(include=['int64', 'float64'])
```

```python
# métricas de tendência central
ct1 = pd.DataFrame(num_attributes.apply(np.mean)).T # média
ct2 = pd.DataFrame(num_attributes.apply(np.median)).T # mediana

# métricas de dispersão
d1 = pd.DataFrame(num_attributes.apply(np.std)).T # desvio-padrão
d2 = pd.DataFrame(num_attributes.apply(min)).T # mínimo
d3 = pd.DataFrame(num_attributes.apply(max)).T # máximo
d4 = pd.DataFrame(num_attributes.apply(lambda x: x.max() - x.min())).T # range
d5 = pd.DataFrame(num_attributes.apply(lambda x: x.skew())).T # assimetria
d6 = pd.DataFrame(num_attributes.apply(lambda x: x.kurtosis())).T # curtose

# concatenar as métricas
metricas = pd.concat([d2, d3, d4, ct1, ct2, d1, d5, d6]).T.reset_index()

# nomear as colunas
metricas.columns = ['atributos', 'min', 'max', 'range', 'media', 'mediana', 
                   'std', 'assimetria', 'curtose']
metricas
```

## MARKETING E CUSTO DA PLATAFORMA

```python
dados['custo_por_cliente_porcentagem(excluindo_marketing)'] = ( 
    dados['custo_por_cliente_(excluindo_marketing)'] / dados['receita_por_cliente']
 * 100) 
dados['custo_por_cliente_porcentagem_apenas_marketing'] = (
    100 - dados['custo_por_cliente_porcentagem(excluindo_marketing)'])
```

```python
# gráficos
plt.plot(dados['data'], dados['custo_por_cliente_porcentagem(excluindo_marketing)'],
         color='dimgray')
plt.plot(dados['data'], dados['custo_por_cliente_porcentagem_apenas_marketing'], 
         color='silver')

# legenda do eixo yy
plt.ylabel('Custo (em milhares de dólares)')

# título
plt.title('CUSTO DE MARKETING E CUSTO DA PLATAFORMA',
             x=0.38, 
             y=1.15,
             color='dimgray',
             weight='bold',
             fontsize=22)

# subtítulo
plt.suptitle('Comparação dos gastos e marketing da Netflix por cliente nos trimestres entre os anos de 2011 e 2021',
             x=0.52, 
             y=0.87,
             color='dimgray',
             fontsize=18)

# texto 1
plt.text(19050, # x
         70, # y
         'Custo da plataforma',
         weight='bold',
         color='dimgray',
         fontsize=16)

# texto 2
plt.text(19050, # x
         30, # y
         'Custo de marketing',
         weight='bold',
         color='silver',
         fontsize=16)

# legenda da figura
plt.gcf().text(0.8, # x
               -0.05, # y
               '\nGráfico 7', # texto
               fontsize=24,
               color='dimgray',
               weight='bold',)

plt.show()
```

Vemos que os gráficos são espelhados, isso se deve ao fato a soma de ambos em um determinado ponto dá 100%. A Netflix ao longo do tempo foi gastando menos com os custos da plataforma e mais com marketing. Isso faz sentido, pois a plataforma investiu em marketing para conseguir novos cliente. No último ano, há um aumento de custo da plataforma, o que pode ser explicado pela pandemia, que deve ter tido um aumento de novos usuários acima da normalidade.

Vamos agora avaliar a porcentagem de assinaturas pagas.

## ASSINATURAS PAGAS

```python
dados['porcentagem_assinaturas_pagas'] = (dados['assinaturas_pagas_no_final_do_periodo']
                                         / dados['total_de_assinaturas_no_final_do_periodo']) * 100
```

```python
plt.plot(dados['data'], dados['porcentagem_assinaturas_pagas'],
         color='dimgray')
plt.ylabel('Porcentagem de usuários (%)')

# título
plt.title('ASSINATURAS PAGAS',
             x=0.17, 
             y=1.15,
             color='dimgray',
             weight='bold',
             fontsize=22)

# subtítulo
plt.suptitle('Porcentagem de assinaturas pagas na Netflix nos trimestres entre os anos de 2011 e 2021',
             x=0.48, 
             y=0.87,
             color='dimgray',
             fontsize=18)

# texto 1
plt.text(19050, # x
         98, # y
         'Custo da plataforma',
         weight='bold',
         color='dimgray',
         fontsize=16)

# legenda da figura
plt.gcf().text(0.8, # x
               -0.05, # y
               '\nGráfico 8', # texto
               fontsize=24,
               color='dimgray',
               weight='bold',)

plt.show()
```

1. Há um elevado cancelamento de assinaturas, sendo esse número maior que o número de novas assinaturas pagas, fazendo com que a porcentagem caia.
2. Há um alto grau de novos assinantes que estão ainda no período gratuito, aumentando o número de assinaturas gratuitas, elevando assim sua porcentagem e abaixando a porcentagem de assinaturas pagas.

Iremos agora olhar a receita de cada trimestre e ano, para ver se a algum padrão o discrepância.

## COMPARAÇÃO DAS RECEITAS AO LONGO DOS ANOS

## COMPARAÇÃO DO CUSTO DA PLATAFORMA E MARKETING AO LONGO DOS ANOS 

```python
# criar variáveis agrupada de ano, mês e custa da receita
aux_gasto_ano = (dados[['ano', 'mes', 'custos_da_receita',
                        'custo_de_marketing']].groupby(['ano', 'mes'])
                   .sum()
                   .reset_index())

# criando figura
plt.figure(figsize=(14, 14))
ax1 = plt.subplot(2, 1, 1)
ax2 = plt.subplot(2, 1, 2)

# gráfico
ax_1 = sns.barplot(x='ano', y='custos_da_receita', hue='mes', data=aux_gasto_ano,
                 color='grey',ax=ax1)
ax_2 = sns.barplot(x='ano', y='custo_de_marketing', hue='mes', data=aux_gasto_ano,
                 color='grey',ax=ax2)
ax1.set_ylabel('Receita (em milhares de dólares)')
ax2.set_ylabel('Receita (em milhares de dólares)')

# subtítulo
plt.suptitle('Gasto da Netflix com marketing e plataforma nos trimestres entre os anos de 2011 e 2021',
             x=0.5, 
             y=1,
             color='dimgray',
             weight='bold',
             fontsize=20)

# rótulos das legendas
labels = ['Março', 'Junho', 'Setembro', 'Dezembro']
h, l = ax_1.get_legend_handles_labels()
ax1.legend(h , labels, title='Meses', loc='upper left')
ax2.legend(h, labels, title='Meses', loc='upper left')

# subtitulo figura

plt.gcf().text(0.45, # x
               0.95, # y
               '\nCusto de plataforma', # texto
               fontsize=18,
               color='dimgray',)

plt.gcf().text(0.45, # x
               0.45, # y
               '\nCusto de Marketing', # texto
               fontsize=18,
               color='dimgray',)

# legenda da figura
plt.gcf().text(0.9, # x
               0.48, # y
               '\nGráfico 10', # texto
               fontsize=24,
               color='dimgray',
               weight='bold',)

plt.gcf().text(0.9, # x
               -0.01, # y
               '\nGráfico 11', # texto
               fontsize=24,
               color='dimgray',
               weight='bold',)

plt.show()
```

* O último trimestre do ano é o de maior gasto da plataforma;
* Há um aumento gradual de investimento na plataforma, com o aumento de custo, exceto no primeiro e segundo semestre de 2021, que estão abaixo em custo do último semestre de 2020.

O `marketing` já é mais imprevisível. O que podemos afimar é que o último trimestre de cada ano é o que mais se tem investimento. Isso pode ser respondido pelos especiais de natal que a Netflix costuma lançar ou as grandes séries que costumam ser lançadas no final do ano, como Squid Game, lançado em 17 de setembro de 2021 e Bridgerton, que foi lançada em 25 de dezembro de 2020 são as duas séries mais vistas no mundo no ano de 2021.

Um possível motivo para o marketing não ter um padrão claro é os grandes lançamentos que a Netflix faz ao longo do ano, não tendo uma época específica para lançar um série ou filme de sucesso. Por exemplo, quando analisamos [os 10 shows de TV e 10 filmes no streming  mais visto do mundo](https://s22.q4cdn.com/959853165/files/doc_financials/2021/q4/FINAL-Q4-21-Shareholder-Letter.pdf), 8 são da Netflix, sendo 5 lançados no último trimestre, 3 no segundo trimestre e 1 no primeiro trimestre (a série Lupin, 7ª colocada, teve duas estreias no ano, sendo da segunda e terceira parte)
